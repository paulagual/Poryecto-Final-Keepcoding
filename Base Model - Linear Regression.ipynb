### Train Test Split

```python
target = 'Price'
X = df.drop(target, axis=1)
y = df[target]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
y_train.describe()
```

```python
y_test.describe()
```

```python
y.describe()
```

### Sklearn Linear Regression

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
X_pred = model.predict(X_train)
```

```python
y_pred = model.predict(X_test)
```

### Results

```python
train = ['Train Base Model',
            round(mean_absolute_error(y_train, X_pred),2),
            round(mean_squared_error(y_train, X_pred, squared=False),2),
            round(r2_score(y_train, X_pred),3)]

test = ['Test Base Model',
           round(mean_absolute_error(y_test, y_pred),2),
           round(mean_squared_error(y_test, y_pred, squared=False),2),
           round(r2_score(y_test, y_pred),3)]

results = pd.DataFrame([train, test])
results.columns=['Model','MAE', 'RMSE','R2']
```

```python
results
```

### Cross Valitation

```python
mae_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='neg_mean_absolute_error')
print("MEAN CV MAE: %.3f" %round(abs(mae_scores.mean()),2))
```

```python
rmse_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='neg_root_mean_squared_error')
print("MEAN CV RMSE: %.3f" %round(abs(rmse_scores.mean()),2))
```

```python
r2_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='r2')
print("MEAN CV MAE: %.3f" %round(abs(r2_scores.mean()),2))
```

### P Values

```python
reg = LinearRegression()
reg.fit(X, y)
beta_hat = [reg.intercept_] + reg.coef_.tolist()
```

```python
n = X.shape[0]
X1 = np.column_stack((np.ones(n), X))
```

```python
sigma_hat = np.sqrt(np.sum(np.square(y - X1@beta_hat)) / (n - X1.shape[1]))
```

```python
beta_cov = np.linalg.inv(X1.T@X1)
```

```python
t_vals = beta_hat / (sigma_hat * np.sqrt(np.diagonal(beta_cov)))
```

```python
p_vals = t.sf(np.abs(t_vals), n-X1.shape[1])*2 
```

```python
p_values = pd.DataFrame(zip(X.columns, p_vals), columns=['Feature', 'P Values']) 
```

### Permutarion Importance

```python
perm_importance = permutation_importance(model, X, y, n_repeats=5,random_state=42)
```

```python
sorted_idx = perm_importance.importances_mean.argsort()
```

```python
plt.barh(X.columns[sorted_idx], perm_importance.importances_mean[sorted_idx])
plt.xlabel("Permutation Importance", size=20)
plt.title('Permutation Importance Base Model', size= 30)
plt.yticks(size=22)
plt.show()
```

```python
fig , ax = plt.subplots()
ax.boxplot(perm_importance.importances[sorted_idx].T,
            vert =False, labels=X.columns[sorted_idx])
ax.set_title( "Permutation Importances Base Model", size=30)
ax.tick_params(axis='both', which='major', labelsize=22)
fig.tight_layout()
plt.show()
```
